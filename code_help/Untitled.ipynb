```python
import tkinter as tk
from tkinter import ttk, messagebox

def recomendar_modelo():
    etiquetas = combo_etiquetas.get()
    objetivo = combo_objetivo.get()
    tipo_dato = combo_dato.get()
    filas = entry_filas.get()
    columnas = entry_columnas.get()
    es_time_series = var_time_series.get()

    try:
        filas = int(filas)
    except:
        filas = 0

    recomendacion = ""

    if etiquetas == "No":
        if tipo_dato == "Datos numéricos":
            recomendacion = "Tu problema parece de Clustering. Prueba con: KMeans."
        else:
            recomendacion = "Puedes usar PCA para reducción de dimensionalidad."
    else:
        if objetivo == "Número (ej: precio, edad)":
            if filas < 1000:
                recomendacion = "Modelo recomendado: Regresión Lineal (simple y fácil de interpretar)."
            else:
                recomendacion = "Modelos recomendados: Random Forest Regressor o XGBoost."
            
            # Si es serie temporal, añadir modelos específicos
            if es_time_series:
                recomendacion += "\nAdemás, como es una serie temporal puedes probar: ARIMA, SARIMA o Prophet."
        
        elif objetivo == "Etiqueta (ej: aprobado/suspenso)":
            if filas < 1000:
                recomendacion = "Modelo recomendado: Regresión Logística."
            else:
                recomendacion = "Modelos recomendados: Random Forest Classifier o SVM."

    if tipo_dato == "Imágenes":
        recomendacion = "Modelo recomendado: CNN (Convolutional Neural Network)."
    elif tipo_dato == "Texto o Secuencias":
        recomendacion = "Modelo recomendado: RNN o LSTM."

    messagebox.showinfo("Recomendación", recomendacion)

# Ventana principal
root = tk.Tk()
root.title("Asistente de Modelos ML")
root.geometry("520x500")
root.config(bg="#f4f4f9")

# Pregunta 1
frame1 = tk.LabelFrame(root, text="¿Tienes etiquetas (valores a predecir)?", bg="#f4f4f9")
frame1.pack(fill="x", padx=10, pady=5)
combo_etiquetas = ttk.Combobox(frame1, values=["Sí", "No"])
combo_etiquetas.set("Sí")
combo_etiquetas.pack(padx=5, pady=5)

# Pregunta 2
frame2 = tk.LabelFrame(root, text="Tipo de resultado que quieres predecir", bg="#f4f4f9")
frame2.pack(fill="x", padx=10, pady=5)
combo_objetivo = ttk.Combobox(frame2, values=["Número (ej: precio, edad)", "Etiqueta (ej: aprobado/suspenso)"])
combo_objetivo.set("Número (ej: precio, edad)")
combo_objetivo.pack(padx=5, pady=5)

# Pregunta 3
frame3 = tk.LabelFrame(root, text="Tipo de datos principales", bg="#f4f4f9")
frame3.pack(fill="x", padx=10, pady=5)
combo_dato = ttk.Combobox(frame3, values=["Datos numéricos", "Imágenes", "Texto o Secuencias"])
combo_dato.set("Datos numéricos")
combo_dato.pack(padx=5, pady=5)

# Pregunta 4 - tamaño dataset
frame4 = tk.LabelFrame(root, text="Tamaño del dataset", bg="#f4f4f9")
frame4.pack(fill="x", padx=10, pady=5)
tk.Label(frame4, text="Número de filas (ej: 5000):", bg="#f4f4f9").pack()
entry_filas = tk.Entry(frame4)
entry_filas.insert(0,"1000")
entry_filas.pack(padx=5, pady=2)
tk.Label(frame4, text="Número de columnas (características):", bg="#f4f4f9").pack()
entry_columnas = tk.Entry(frame4)
entry_columnas.insert(0,"10")
entry_columnas.pack(padx=5, pady=2)

# Pregunta 5 - serie temporal
frame5 = tk.LabelFrame(root, text="¿Tus datos son de tipo serie temporal (fechas, tiempo)?", bg="#f4f4f9")
frame5.pack(fill="x", padx=10, pady=5)
var_time_series = tk.BooleanVar()
chk_time_series = tk.Checkbutton(frame5, text="Sí, son serie temporal", variable=var_time_series, bg="#f4f4f9")
chk_time_series.pack(padx=5, pady=5)

# Botón recomendar
btn = tk.Button(root, text="Recomendar Modelo", command=recomendar_modelo,
                bg="#0d253f", fg="white", font=("Arial", 12, "bold"))
btn.pack(pady=20)

root.mainloop()
```
