```python
import numpy as np
import matplotlib.pyplot as plt

# Parámetros de la distribución
media = 3
desviacion = 1.5
n = 10000

# Generar distribución gaussiana
datos = np.random.normal(media, desviacion, n)

# Calcular el rango del 95% (regla de 2 sigmas)
lim_inf = media - 2 * desviacion
lim_sup = media + 2 * desviacion

print(f"El 95% de los valores se encuentran en el rango: {lim_inf:.2f} < x < {lim_sup:.2f}")

# Dibujar histograma
plt.hist(datos, bins=50, color="skyblue", edgecolor="black", density=True)
plt.axvline(lim_inf, color="red", linestyle="--", label=f"Límite inferior ({lim_inf:.2f})")
plt.axvline(lim_sup, color="green", linestyle="--", label=f"Límite superior ({lim_sup:.2f})")
plt.axvline(media, color="black", linestyle="-", label=f"Media ({media})")
plt.legend()
plt.title("Distribución Gaussiana (N=10,000)")
plt.xlabel("Valor")
plt.ylabel("Densidad")
plt.show()
```
