```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)

df_abc = pd.read_csv(
    "ABC_fam.csv",
    encoding="ISO-8859-1",
    delimiter=","
)

df_abc
```

```python


# Suponiendo que ya tienes cargado tu df (ej: df = pd.read_csv(...))

# Crear columna 'ID' desde 1 hasta número de filas
df_abc.insert(0, "ID", range(1, len(df_abc) + 1))



df_abc.head(20)
```

```python
df_filtrado = df_abc[(df_abc["ID"] >= 336) & (df_abc["ID"] <= 403)]
df_filtrado
```

```python




print(df_filtrado[["ID", "Descripcion", "Unidades"]])

# Ordenar de mayor a menor por 'Unidades'
df_ordenado = df_filtrado.sort_values(by="Unidades", ascending=False)
df_ordenado = df_ordenado[~df_ordenado["ID"].isin([342, 343, 338, 351, 398, 344])]
#df_ordenado[["ID","Descripcion", "Unidades"]].head(31)
print(df_ordenado[df_ordenado["ID"].isin([343, 344, 351])])
df_ordenado[["ID","Descripcion", "Unidades"]].head(50)
```

```python
import matplotlib.pyplot as plt

# Tomamos top 20 para graficar
top = df_ordenado.head(20)

# Gráfico de barras horizontal
plt.figure(figsize=(10,8))
plt.barh(top["Descripcion"], top["Unidades"], color="skyblue")
plt.xlabel("Unidades")
plt.ylabel("Descripción")
plt.title("Top 20 productos por unidades (df_ordenado)")
plt.gca().invert_yaxis()  # mayor arriba
plt.tight_layout()
plt.show()

# Gráfico de barras vertical por ID
plt.figure(figsize=(12,6))
plt.bar(top["ID"].astype(str), top["Unidades"], color="orange")
plt.xlabel("ID")
plt.ylabel("Unidades")
plt.title("Top 20 productos por ID (df_ordenado)")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
